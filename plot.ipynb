```python
import matplotlib.pyplot as plt

def plot_results(results):
    recalls = [r[0] for r in results]
    avg_calculations = [r[1] for r in results]
    ef_values = [r[2] for r in results]

    plt.figure(figsize=(8,5))
    plt.plot(recalls, avg_calculations, color="blue")

    for i, ef in enumerate(ef_values):
        plt.annotate(f"ef={int(ef)}", (recalls[i], avg_calculations[i]), textcoords="offset points", xytext=(0,5), ha="center")

    plt.xlabel("Recall")
    plt.ylabel("Average calculations")
    plt.title("Recall vs Average calculations for different ef")
    plt.grid(True)
    plt.show()
    # plt.savefig("results.png")

if __name__=="__main__":
    results = []
    with open("results.txt", "r") as f:
        next(f)
        for line in f:
            recall, avg_calc, ef = map(float, line.strip().split(","))
            results.append((recall, avg_calc, int(ef)))

    plot_results(results)
```
